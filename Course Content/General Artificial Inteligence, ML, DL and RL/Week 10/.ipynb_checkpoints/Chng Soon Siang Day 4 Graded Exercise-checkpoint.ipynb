# Define the Linear Transformation Function

```python
img1 = img[1:4,1:4,:]
print("*")
print(img1)
np.take(img1, [[0,1,2],[0,1,2]], mode = 'wrap')
```

```python
def linear_transformation(src, a):
    M, N, O = src.shape
    points = np.mgrid[0:N, 0:M].reshape((2, M*N))
    new_points = np.linalg.inv(a).dot(points).round().astype(int)
#     print(new_points)
    x, y = new_points.reshape((2, M, N), order='F')
    
    x = np.mod(x, N)
    y = np.mod(y, M)
    
#     print(indices)
    return src[y,x,:]
```

# Transformation 2
## Shearing about the x-axis with a vertical displacement of +y/2

```python
img = mpimg.imread('img.jpg')
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
img = rgb2gray(img)
imgplot = plt.imshow(img)
```

```python
def linear_transformation(src, a): # Original function from the internet.
    M, N = src.shape
    points = np.mgrid[0:N, 0:M].reshape((2, M*N))
    new_points = np.linalg.inv(a).dot(points).round().astype(int)
    x, y = new_points.reshape((2, M, N), order='F')
    print(x)
    print(x.shape)
    indices = x + N*y
    return np.take(src, indices, mode='wrap')
```

```python
a = np.array([[1, 1],
              [0, 1.5]])
dst = linear_transformation(img, a)
plt.imshow(dst)
plt.show()
```
