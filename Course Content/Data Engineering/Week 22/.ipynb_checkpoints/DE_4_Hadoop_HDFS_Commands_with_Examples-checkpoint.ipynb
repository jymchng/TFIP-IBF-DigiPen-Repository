# Hadoop HDFS Commands

Before starting with the HDFS command, we have to start the Hadoop services. To start the Hadoop services do the following:

1. Open cmd in Administrative mode and 
2. move to “C:\Hadoop\hadoop-3.1.0\sbin” and 
3. start cluster start-dfs.sh

#### Let us now start with the HDFS commands.

**1. version**

```python
#C:\Hadoop\hadoop-3.1.0\sbin>hadoop version
hadoop version
```

**2. mkdir**

```python
Syntax: hadoop fs –mkdir /path/directory_name
```

In this example, we are trying to create a **testData** named directory in HDFS using the **mkdir** command.

```python
Example: hadoop fs –mkdir /testData
```

This command creates the directory in HDFS if it does not already exist.  

**Note:** If the directory already exists in HDFS, then we will get an error message that file already exists.  

Use **hadoop fs mkdir -p /path/directoryname**, so not to fail even if directory exists.  

**3. ls**

```python
Syntax: hadoop fs -ls /path
```

the **ls** command to enlist the files and directories present in HDFS

```python
Example: hadoop fs -ls /
```

Command ls displays a list of the contents of a directory specified in the path provided by the user. It shows the name, permissions, owner, size, and modification date for each file or directories in the specified directory.

```python
hadoop fs -ls -R / 
```

This Hadoop fs command behaves like -ls, but recursively displays entries in all subdirectories of a path.

**4. put**

```python
Syntax: haoop fs -put <localsrc> <dest>
```

Here in this example, we are trying to copy localfile1 of the local file system to the Hadoop filesystem.

```python
Example:
    
    hadoop fs -put ~/localTestFile1 /testData/fileFromLocal
```

The Hadoop fs shell command **put** is similar to the **copyFromLocal**, which copies files or directory from the local filesystem to the destination in the Hadoop filesystem.

**5. copyFromLocal**

```python
Syntax: 
    hadoop fs -copyFromLocal <localsrc> <hdfs destination>
```

Here in the below example, we are trying to copy the ‘test1’ file present in the local file system to the testData directory of Hadoop.

```python
Example:
    hadoop fs -copyFromLocal ~/localTestFile2 /testData/fileFromLocal
```

This command copies the file from the local file system to HDFS.

**6. get**

```python
Syntax:
    hadoop fs -get <src> <localdest>
```

In this example, we are trying to copy the ‘testfile’ of the hadoop filesystem to the local file system.

The Hadoop fs shell command **get** copies the file or directory from the **Hadoop file system to the local file system**.

```python
Example: 
    hadoop fs -get /testfile ~/copyFromHadoop
```

**7. copyToLocal**

```python
Syntax: 
    hadoop fs -copyToLocal <hdfs source> <localdst>
```

Here in this example, we are trying to copy the **sample** file present in the **testData** directory of HDFS to the local file system.

```python
Example:
    hadoop fs -copyToLocal /testData/sample  ~/copySample
```

copyToLocal command copies the file from HDFS to the local file system.

**8. cat**

```python
Syntax: 
    hadoop fs –cat /path_to_file_in_hdfs
```

Here in this example, we are using the **cat** command to display the content of the **sample** file present in **testData directory of HDFS**.

```python
Example:
    hadoop fs -cat /testData/sample
```

The cat command reads the file in HDFS and displays the content of the file on console or stdout.

**9. mv**

```python
Syntax:
    hadoop fs -mv <src> <dest>
```

In this example, we have a directory **DR1** in **HDFS**. We are using **mv command** to move the **DR1 directory** to the **testData** directory in **HDFS**.

```python
Example:
    hadoop fs -mv /DR1 /testData
```

The HDFS mv command moves the files or directories from the source to a destination within HDFS.

**10. cp**

```python
hadoop fs -cp <src> <dest>
```

In the below example we are copying the **file1** present in testData1 directory in HDFS to the testData directory of HDFS.

```python
Example:
    hadoop fs -cp /testData1/file1 /testData
```

The **cp** command copies a file from one directory to another directory within the HDFS.
