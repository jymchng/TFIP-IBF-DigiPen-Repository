### (DONE) Question 2  
  
Assume that each of the expressions below gives the execution time of an algorithm for solving a problem of size $n$. Select the dominant term(s) having the steepest increase in $n$ and specify the big O notation of each algorithm.

# BIG LP: O(n) notation is not in the format of O(n) = function of n but it is in the form of O(function(n)).
# e.g. O(n$\log{n}$), O($n^2$) etc.
 
 

**(a)** $5+0.001n^3+0.025n$  Dominant term: $0.001n^3$, O(n) = $n^3$ (Correct but wrong format, correct: O($n^3$)
  
**(b)** $500n + 100n^{1.5} + 50n \log_{10}{n}$  Dominant term: $100n^{1.5}$, O(n) = $n^{1.5}$
  
**(c)** $0.3n + 5n^{1.5} + 2.5n^{1.75}$  Dominant term: $2.5n^{1.75}$, O(n) = $n^{1.75}$
  
**(d)** $n^2\log_2{n}+n(\log_2{n})^2$  Dominant term: $n^2\log_2{n}$, O(n) = $n^2$ 

(LP: the $\log_2{n}$ cannot be ignored) 

(Correct: O($n^2\log_2$))
  
**(e)** $n\log_3{n}+n\log_2{n}$  Dominant term: $n\log_3{n}$, O(n) = $n\log_3{n}$ 

(LP: the logarithmic base can be neglected, $n\log_3{n}$ is equivalent to $n\log{n}$ in O(n) notation.)

(Correct: $O(n\log{n})$)
  
**(f)** $3\log_8{n}+\log_2\log_2\log_2{n}$  Dominant term: $3\log_8{n}$, O(n) = $log_8{n}$

(Correct: O($log_{n}$))
  
**(g)** $100n+0.01n^2$  Dominant term: $0.01n^2$, O(n) = $n^2$
  
**(h)** $0.01n+100n^2$  Dominant term: $100n^2$, O(n) = $100n^2$ (Correct but wrong format)
  
**(i)** $2n+n^{0.5}+0.5n^{1.25}$  Dominant term: $0.5n^{1.25}$, O(n) = $n^{1.25}$ (Correct but wrong format)
  
**(j)** $0.01n\log_2{n}+n(\log_2{n})^2$  Dominant term: $0.01n\log_2{n}$, O(n) = $n\log_2{n}$

### WRONG! $n(\log_2{n})^2$ > $0.01n\log_2{n}$ for n > some number

Correct Answer: Dominant term: $n(\log_2{n})^2$, O($n(\log{n})^2$)
  
**(k)** $100n\log_3{n}+n^3+100n$  Dominant term: $n^3$, O(n) = $n^3$ (Correct but wrong format)
  
**(l)** $0.03\log_4{n}+\log_2\log_2{n}$  Dominant term: $0.03\log_4{n}$, O(n) = $log_4{n}$ (Correct but wrong format)

### Question 4 (DONE)
  
Assume that the array `a` contains $n$ values, that the function `randomValue` takes constant number c of computational steps to produce each output value, and that the method `goodSort` takes $n \log{n}$ computational steps to sort the array. Determine the big O complexity for the following fragments of code taking into account only the above computational steps:

```python
for i in range(n):
    for j in range(n):
        a[j] = randomValue(i)
    goodSort(a)
```

### Question 4 Answer

$k = n*(c*n+n\log{n})$

$k = c*n^2+n^2*\log{n}$

O($n^2\log{n}$)

### Question 5(a) Answer

````python
 
from math import log2, factorial
import matplotlib.pyplot as plt


x = []
first = [2]
ii = [i for i in range (1, 11)]
iii = []
iv = []
v = []
vi = []
vii = []


for i in range (1, 11):
    x.append(i)
    first.append(2)
    iii.append(i**2)
    iv.append(log2(i))
    v.append(i*log2(i))
    vi.append(2**i)
    vii.append(factorial(i))


plt.figure(figsize=(20,10))
axes = plt.gca()
axes.plot(i, label='f(n)=2')
axes.plot(ii, label='f(n)=n')
axes.plot(iii, label='f(n)=n^2')
axes.plot(iv, label='f(n)=log2n')
axes.plot(v, label='f(n)=nlog2n')
axes.plot(vi, label='f(n)=2^n')
axes.plot(vii, label='f(n)=n!')
axes.legend()
axes.set_ylim([0,40])
plt.show()
````
