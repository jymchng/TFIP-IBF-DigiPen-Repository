**Q1)** *From exercise on Conditionals Q1*    
Write a function called `find_largest_num()` that has 3 input arguments (`num1`, `num2` and `num3`) and return the largest number among the 3 input arguments. The 3 input integers will be provided by the user. Remember to include the function docstring.

```python
def find_largest_num(num1, num2, num3):
    '''
    Function to find the largest number among 3 numbers
    Input:
        num1, num2, num3 - 3 integer numbers
    Return:
        largest integer numbers among the 3
    '''
    max_num = num1
    if max_num < num2:
        max_num = num2
    if max_num < num3:
        max_num = num3
    
    return max_num

num1 = int(input("Enter a number 1: "))
num2 = int(input("Enter a number 2: "))
num3 = int(input("Enter a number 3: "))
print('The maximum number is', find_largest_num(num1, num2, num3))
```

**Q2)** *From exercise on Conditionals Q5*    
Write a function called `cal_quadratic_eqn()` that has 3 input arguments (`a`, `b` and `c`) and return 3 outputs (a boolean and 2 floats). 

$$
\frac{-b\pm \sqrt{b^{2}-4ac}}{2a}
$$

The function must clearly reflect the following:
* calculates the roots return 3 values
* `True` to indicate a successful calculation and `False` for unsuccessful calculation
* 2 roots (regardless of whether they are same or different) upon successful calculation
* return 2 zeroes when the roots are complex
* check that input `a` cannot be 0. If it is, return `False` and 2 zeroes 

The program must clearly reflect the following:
* print nice messages for the users

**Program Output Example**
<pre>
Enter a number for variable a: 5
Enter a number for variable b: 6
Enter a number for variable c: 1

The first root is  -0.2
The second root is  -1.0
</pre>

```python
from math import sqrt

def cal_quadratic_eqn(a,b,c):
    '''
    Function to calculate the Qudratic equation given the inputs
    Input:
        a, b, c - 3 numbers
    Return:
        boolean - to indicate the successfulness of the calculation
        2 roots - float/integer numbers indicating the calculated roots
    '''
    if a == 0:
        return False, 0, 0
    
    discriminant = b**2 - 4*a*c

    if discriminant == 0:
        roots = -b/(2*a)
        return True, roots, roots
    elif discriminant > 0:
        root1 = (-b + sqrt(discriminant))/(2*a)
        root2 = (-b - sqrt(discriminant))/(2*a)
        return True, root1, root2
    else:
        return True, 0, 0

    
a = int(input("Enter a number for variable a: "))
b = int(input("Enter a number for variable b: "))
c = int(input("Enter a number for variable c: "))

success, root1, root2 = cal_quadratic_eqn(a,b,c)

print()
if success:
    print("The first root is ", root1)
    print("The second root is ", root2)
else:
    print("Variable a cannot be 0")
```

**Q5)** *From exercise on Loops Q5*    
Write a function called `cpf_calculator()` that calculates the CPF contribution for an employee based on their monthly salary and age. The function is to return the employee's and employer's CPF contribution.

The program (not the function) will display the employee's, employer's and total CPF contribution and only terminate if the letter `q` (case-insensitive) is entered in the prompt for monthly salary.

The table below shows the rates:

| Employee's Age | Employee Rate (%) | Employer Rate (%) |
|:---:|:---:|:---:|
| 55 and Above | 20 | 17 |
| above 55 to 60 | 13 | 13 |
| above 60 to 65 | 7.5 | 9 |
| above 65 | 5 | 7.5 |

**Program Output Example**
<pre>
Enter monthly salary: $3000
Enter age: 30

The employee's contribution is: $ 600.0
The employer's contribution is: $ 510.00000000000006
The total contribution is: $ 1110.0
</pre>

```python
def cpf_calculator(sal, age):
    '''
    Function to calculate the CPf contribution for an employee
    Input:
        sal - a float number, monthly salary
        age - an integer number, age of the employee
    Return:
        2 floating point numbers of the employee's and employer's CPF contribution
    '''
    employee_contri = 0
    employer_contri = 0
    
    if age <= 55:
        employee_contri = sal * 0.2
        employer_contri = sal * 0.17
    elif age > 55 and age <= 60:
        employee_contri = sal * 0.13
        employer_contri = sal * 0.13
    elif age > 60 and age <= 65:
        employee_contri = sal * 0.075
        employer_contri = sal * 0.09
    else:
        employee_contri = sal * 0.05
        employer_contri = sal * 0.075
    
    return employee_contri, employer_contri


while True:
    sal = input("Enter monthly salary: $")
    if sal.lower() == 'q':
        break
    sal = float(sal)
    age = int(input("Enter age: "))
    
    employee, employer = cpf_calculator(sal, age)
    total_contri = employee + employer
    
    print()
    print("The employee's contribution is: $", employee)
    print("The employer's contribution is: $", employer)
    print("The total contribution is: $", total_contri)
    print()
```

**Q6)** Write a function called `is_perfect()` that accepts a **single** integer and returns a boolean. The function is to check if the given number is a perfect number or not. 

>From Britannica:    
Perfect number, a positive integer that is equal to the sum of its proper divisors. The smallest perfect number is 6, which is the sum of 1, 2, and 3. Other perfect numbers are 28, 496, and 8,128.

For Example:     
6 is a perfect number because 1,2, and 3 are its proper divisors. 
* 1 + 2 + 3 = 6
* 1 + 2 + 4 + 7 + 14 = 28

The program is to get the number from user input before calling the function to check. A nice message is to be printed to inform the user of the result. **Assume that the user will always enter an integer.**

**Program Output Example**
<pre>
Enter a number:  8
Is the number 8 perfect? False

Enter a number:  496
Is the number 496 perfect? True
</pre>

```python
def is_perfect(n):
    '''
    Function to check of the given number is a perfect number
    Input:
        n - number to check
    Return
         boolean
    '''
    sum = 0
    for x in range(1, n):
        if n % x == 0:
            sum += x
    return sum == n


user_int = int(input("Enter a number: "))
print("Is the number", user_int, "perfect?", is_perfect(user_int))
```

The user input will provide the number of rows for Pattern A **but** Pattern B, C and D have incremental rows based on the following rules:

* number of rows for Pattern B is 2 more than pattern A
* number of rows for Pattern C is 2 times than that of pattern A
* number of rows for Pattern D is 2 times + 3 more than that of pattern A

```python
def pattern_A(size):
    '''
    Function to print Pattern A
    Input:
        size - integer number, number of rows for Pattern A
    '''
    # Pattern A ---------------------------------------
    print('\nPattern A')
    for row in range(1, size+1):
        for col in range(1, row+1):
            print(col, end=' ')
        print()

        
def pattern_B(size):
    '''
    Function to print Pattern B
    Input:
        size - integer number, number of rows for Pattern B
    '''
    # Pattern B ---------------------------------------
    print('\nPattern B')
    for row in range(1, size+1):
        for col in range(size, 0, -1):
            if (col <= row):
                print(col, end=' ')
            else:
                print(' ', end=' ')
        print()

        
def pattern_C(size):
    '''
    Function to print Pattern C
    Input:
        size - integer number, number of rows for Pattern C
    '''
    # Pattern C ---------------------------------------
    print('\nPattern C')
    for row in range(size):
        for col in range(1, size+1):
            if col > row:
                print(col-row, end=' ')
            else:
                print(' ', end=' ')
        print()

def pattern_D(size):
    '''
    Function to print Pattern D
    Input:
        size - integer number, number of rows for Pattern D
    '''
    # Pattern D ---------------------------------------
    print('\nPattern D')
    for row in range(size):
        for col in range (size,0, -1):
            if col > row:
                print(col-row, end=' ')
            else:
                print(' ', end=' ')
        print()

    
size = int(input('Enter an integer for size of Pattern A: '))
pattern_A(size)
pattern_B(size + 2)
pattern_C(size * 2)
pattern_D(size * 2 + 3)
```
