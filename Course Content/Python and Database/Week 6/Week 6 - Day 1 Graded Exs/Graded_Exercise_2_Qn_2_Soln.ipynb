```python
import re

class Calculator():
    '''
    Calculator that does element-wise calculations with lists and floats
    '''
    def __init__(self):
        '''Constructor'''
        self.result = None
    
    def add(self, val1, val2):
        '''
        Function to add two values.
        Values are either integers or lists of same length numerals
        Returns either an integer or a list of numerals
        '''
        # to check the datatype of the value
        isList_v1 = isinstance(val1, list)
        isList_v2 = isinstance(val2, list)

        # Scenario 1: element-wise calculation for 2 lists, returns a list
        if isList_v1 and isList_v2:
            self.result = []  # changing the variable to a list
            for v1, v2 in zip(val1, val2):
                self.result.append(v1 + v2)

        # Scenario 2: element-wise calculation for a list and 
        #             a single float number, returns a list
        elif isList_v1 or isList_v2:
            self.result = []  # changing the variable to a list
            if isList_v1:
                for v1 in val1:
                    self.result.append(v1 + val2)
            else:
                for v2 in val2:
                    self.result.append(v2 + val1)

        # Scenario 3: calculation for 2 single float numbers, return 
        #             single float number
        else:
            self.result = val1 + val2

        return self.result
    
    def divide(self, val1, val2):
        '''
        Function to divide two values.
        Values are either integers or lists of same length numerals
        Returns either an integer or a list of numerals
        '''
        # to check the datatype of the value
        isList_v1 = isinstance(val1, list)
        isList_v2 = isinstance(val2, list)

        # Scenario 1: element-wise calculation for 2 lists, returns a list
        if isList_v1 and isList_v2:
            self.result = []  # changing the variable to a list
            for v1, v2 in zip(val1, val2):
                try:
                    self.result.append(round(v2 / v1, 2))
                except ZeroDivisionError:
                    self.result.append('N.A')

        # Scenario 2: element-wise calculation for a list and 
        #             a single float number, returns a list
        elif isList_v1 or isList_v2:
            self.result = []  # changing the variable to a list
            if isList_v1:
                for v1 in val1:
                    try:
                        self.result.append(round(v1 / val2, 2))
                    except ZeroDivisionError:
                        self.result.append('N.A')
            else:
                for v2 in val2:
                    try:
                        self.result.append(round(v2 / val1, 2))
                    except ZeroDivisionError:
                        self.result.append('N.A')

        # Scenario 3: calculation for 2 single float numbers, return 
        #             single float number
        else:
            try:
                self.result = round(val2 / val1, 2)
            except ZeroDivisionError:
                self.result = 'N.A'
        
        return self.result
    
    @staticmethod
    def extract_input(val):
        '''
        Function to extract numbers from strings.
        '''
        # pattern = r'[\d.,+-]+[^\W]|\d'
        pattern = r'[-]?\d+\.?\d*'
        
        result = [re.findall(pattern, item) for item in val.split(',')]
        
        if not any(result):
            return None
        return ','.join([i[0] for i in result if len(i) == 1])
    
    @staticmethod
    def convert_str_to_numerical(str_val):
        '''
        Function to convert a valid string to a list of numerals or integer.
        Returns a list of numerals
        '''
        # 1st check: To see of the input can be converted to a float
        try:
            return float(str_val)
        except ValueError:
            result = []
            # 2nd check: check if it is a string of comma separated numerals
            for i in str_val.split(","):
                try:
                    result.append(float(i))
                except ValueError:
                    pass
            
            if not result:
                return None
            return result
```

```python
# option 1 without `validate_input` function

cal = Calculator()

# always True because the calculator is supposed to run until 'q' (case-insensitive) is entered
while(True):
    
    # requesting for 'type of operation' from the user
    type_op = input("Enter operation to perform: ")
    # end the loop if 'quit' has been entered
    if type_op.lower() == 'q':
        print('Bye Bye')
        break
    
    # checks that the 'type of operation' input is correct
    if ((type_op != '+') and (type_op != '/')):
            print("Please enter a correct operator.")
    else:
        # request for the 2 values after a valid 'type of operation'
        val_1 = input("Enter 1st value: ")
        val_2 = input("Enter 2nd value: ")
        
    #-------------------- Changes here -------------------------------------------
        # attempt to convert the values
        val_1 = Calculator.convert_str_to_numerical(val_1)
        val_2 = Calculator.convert_str_to_numerical(val_2)
        
#         print(val_1)
#         print(val_2)
        result = None
        # checks to see if the conversion is successful
        if val_1 is not None and val_2 is not None:
            if type_op == '+':
                result = cal.add(val_1, val_2)
            elif type_op == '/':
                result = cal.divide(val_1, val_2)
        else:
            print('Operation cannot be performed because of invalid operands.')
        
        if result is not None:
            print(f'Results of {val_1} {type_op} {val_2} = {result}')

    # for formating purposes
    print()
```
